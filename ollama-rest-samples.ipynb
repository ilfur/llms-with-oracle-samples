# Defining function for easier REST service calls

```python
def call_chat_service(enquiry):
    url = 'http://ollama.meinnetzwerk.com/api/chat'
    headers = {'Content-Type': 'application/json', 'Accept':'*/*'}
    response = requests.post(url, json=enquiry, headers=headers, verify=False)
    json_obj = json.loads(response.text)
    pretty= json.dumps(json_obj, indent=4)
    print(pretty)

def print_chat_service_content(enquiry):
    url = 'http://ollama.meinnetzwerk.com/api/chat'
    headers = {'Content-Type': 'application/json', 'Accept':'*/*'}
    response = requests.post(url, json=enquiry, headers=headers, verify=False)
    json_obj = json.loads(response.text)
    json_obj = (json_obj["message"])
    print (json_obj["content"])

def call_completion_service(enquiry):
    url = 'http://ollama.meinnetzwerk.com/api/generate'
    headers = {'Content-Type': 'application/json', 'Accept':'*/*'}
    response = requests.post(url, json=enquiry, headers=headers, verify=False)
    json_obj = json.loads(response.text)
    pretty= json.dumps(json_obj, indent=4)
    print(pretty)
```

```python
question = "please generate some python code to create an audio transcription from a file using the whisper api"
enquiry = {"model": "codellama","messages": [{"role": "user",   "content": question }],
	       "stream": False, "options": {"use_mmap": True,"use_mlock": True,"num_thread": 4}}
           
print_chat_service_content(enquiry)
```

## asking deepseek a medical question

```python
question = "can men get breast cancer?"
enquiry = {"model": "deepseek-r1:1.5b","messages": [{"role": "system", "content": "you are a lightly arrogant medical doctor. your answer should be short and no more than 5 sentences."}, 
                                                    {"role": "user",   "content": question }],
	       "stream": False, "options": {"use_mmap": True,"use_mlock": True,"num_thread": 4}}
           
call_chat_service(enquiry)
```

## Structured output, requires newer ollama version
